### Import Library

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import re
import random
```

### Data

```python
data = pd.read_csv("data_clean2.csv")
data
```

```python
data = data.astype({'user_name' : 'string'})
data = data.astype({'text' : 'string'})
data = data.astype({'label' : 'category'})
data = data.astype({'lowercase' : 'string'})
data = data.astype({'remove' : 'string'})
data = data.astype({'tweet_clean' : 'string'})
data = data.astype({'text_clean' : 'string'})

data.dtypes
```

### TFIDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tf = TfidfVectorizer(analyzer='word', token_pattern=r"\b\w+\b", min_df=0)
text_tf = tf.fit_transform(data['tweet_clean'])
#cos_sim = cosine_similarity(text_tf, text_tf)
print ("TF: \n", text_tf)

terms = tf.get_feature_names()
print ("TERMS: \n", terms)

df = pd.DataFrame(text_tf.T.todense(), index=terms) 
#df.sort_values(by=["tfidf"],ascending=False)
df.head()

#h_tfidf = TfidfTransformer()
#x_tfidf = h_tfidf.fit_transform(text_tf)
#print(x_tfidf)
#X = data_clean.text
#Y = data_clean.label
```

### Split Data 95% 5%

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(text_tf, data['label'], test_size=0.05, random_state = 42)
#print ("Data training X: \n", X_train)
print ("Data training Y: \n", y_train)
#print ("Data testing X: \n", X_test)
print ("Data testing Y: \n", y_test)
```

#### Implementasi Naive Bayes & Confusion Matrix 

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

clf = MultinomialNB().fit(X_train, y_train)
training = clf.predict(X_train)
predicted = clf.predict(X_test)

print ("Hasil implementasi data training",training)
print ("Hasil implementasi data testing", predicted)


print ("MultinomialNB Accuracy:", accuracy_score(y_test, predicted))
print ("MultinomialNB Precisio:", precision_score(y_test, predicted, average = "binary", pos_label = "negatif"))
print ("MultinomialNB Recall:", recall_score(y_test, predicted, average = "binary", pos_label = "negatif"))
print ("MultinomialNB f1_score:", f1_score(y_test, predicted, average = "binary", pos_label = "negatif"))

print (f' confusion matrix:\n {confusion_matrix(y_test, predicted)}')
print (' ========================================\n')
print (classification_report(y_test, predicted, zero_division=0))
```
